```python
import json
```

```python
ds = json.loads(open('./anno_test.json', 'r').read())
print(ds)
agents = ['agent_1', 'agent_2']
ag = 0
new_json = {}

for key in ds:
    new_inst = {}
    content = []

    for item in ds[key]:
        content.append({'message':item[0],'agent':agents[ag]})

        if ag == 0:
            ag = 1
        else:
            ag = 0
    new_inst['article_url'] = len(new_json.keys())
    new_inst['content'] = content
    new_json[key] = new_inst
```

```python
print(new_json)
with open('./tc_anno_test.json', 'w') as fp:
    json.dump(new_json, fp)
```
