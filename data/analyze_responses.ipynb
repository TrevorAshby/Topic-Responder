script to expand responder data

```python
# split data to allow for statement / question to be fine tuned upon.
wfile = open('responder_train_data_expanded.txt', 'w')
for i in range(len(df)):
    print(df.iloc[i])

    second = ''
    # check the splitting 
    #try:
    if '?' in df.iloc[i][1] and '.' in df.iloc[i][1]:
        lastpunct = df.iloc[i][1].rfind('.')
        first = df.iloc[i][1][:lastpunct+1]
        second = df.iloc[i][1][lastpunct+1:]
    elif '?' in df.iloc[i][1] and '!' in df.iloc[i][1]:
        lastpunct = df.iloc[i][1].rfind('!')
        first = df.iloc[i][1][:lastpunct+1]
        second = df.iloc[i][1][lastpunct+1:]
    #except:
    #    first = df.iloc[i][1]

    if second != '':
        wfile.write(f'{df.iloc[i][0]}|{first}\n')
        wfile.write(f'{df.iloc[i][0]}|{second}\n')
    else:
        wfile.write(f'{df.iloc[i][0]}|{first}\n')
wfile.close()
```
